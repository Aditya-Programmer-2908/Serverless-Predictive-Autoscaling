````python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import joblib

# ------------------------------
# Load and Inspect Dataset
# ------------------------------
# Purpose: Load the Azure Functions dataset and verify its structure.
DATA_PATH = '/kaggle/input/azure-functions/azurefunctions-accesses-2020.csv'
df = pd.read_csv(DATA_PATH)
print("Dataset columns:", df.columns.tolist())
print("Sample data:\n", df.head())

# ------------------------------
# Anomaly Detection
# ------------------------------
# Purpose: Detect and filter anomalies in BlobBytes and invocation counts to improve model robustness.
def detect_anomalies(df_func, columns=['count', 'BlobBytes'], z_threshold=3):
    """
    Detect anomalies using z-score for specified columns.
    Returns: DataFrame with anomalies removed.
    """
    df_clean = df_func.copy()
    for col in columns:
        z_scores = (df_clean[col] - df_clean[col].mean()) / df_clean[col].std()
        df_clean = df_clean[abs(z_scores) < z_threshold]
    print(f"Removed {len(df_func) - len(df_clean)} anomalies from {columns}")
    return df_clean

# ------------------------------
# Preprocess Data with Enhanced Features
# ------------------------------
# Purpose: Preprocess data for a specific AnonAppName, including temporal and categorical features.
def preprocess_data(df, function_id, sequence_length=60, target_columns=['count', 'BlobBytes']):
    """
    Preprocess data for a specific AnonAppName.
    Args:
        df: DataFrame with Azure Functions data
        function_id: AnonAppName (e.g., '9gti3olh')
        sequence_length: Number of time steps (e.g., 60 minutes)
        target_columns: Columns to predict (e.g., ['count', 'BlobBytes'])
    Returns:
        X_train, y_train, X_test, y_test, scaler
    """
    df_func = df[df['AnonAppName'] == function_id].sort_values(by='Timestamp')
    df_func['Timestamp'] = pd.to_datetime(df_func['Timestamp'], unit='ms')
    
    # Extract temporal features
    df_func['Hour'] = df_func['Timestamp'].dt.hour
    df_func['Day'] = df_func['Timestamp'].dt.dayofweek
    df_func['Week'] = df_func['Timestamp'].dt.isocalendar().week
    
    # Aggregate by minute
    df_agg = df_func.set_index('Timestamp').resample('1min').agg({
        'AnonFunctionInvocationId': 'count',  # Invocations
        'BlobBytes': 'sum',                   # Total storage
        'Read': 'sum',                        # Read operations
        'Write': 'sum',                       # Write operations
        'Hour': 'first',                      # Temporal features
        'Day': 'first',
        'Week': 'first'
    }).rename(columns={'AnonFunctionInvocationId': 'count'}).fillna(0).reset_index()
    
    # Anomaly detection
    df_agg = detect_anomalies(df_agg, columns=['count', 'BlobBytes'])
    
    # Scale features
    features = ['count', 'BlobBytes', 'Read', 'Write', 'Hour', 'Day', 'Week']
    scaler = MinMaxScaler()
    data_scaled = scaler.fit_transform(df_agg[features])
    
    # Train/test split (80/20)
    train_size = int(len(data_scaled) * 0.8)
    train_data, test_data = data_scaled[:train_size], data_scaled[train_size:]
    
    # Create sequences
    def create_sequences(data, seq_length, target_indices=[0, 1]):
        X, y = [], []
        for i in range(len(data) - seq_length):
            X.append(data[i:i + seq_length])
            y.append(data[i + seq_length, target_indices])  # Predict count and BlobBytes
        return np.array(X), np.array(y)
    
    X_train, y_train = create_sequences(train_data, sequence_length)
    X_test, y_test = create_sequences(test_data, sequence_length)
    
    return X_train, y_train, X_test, y_test, scaler

# ------------------------------
# Custom Dataset
# ------------------------------
# Purpose: Create a PyTorch Dataset for efficient batching.
class WorkloadDataset(torch.utils.data.Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)
    
    def __len__(self):
        return len(self.X)
    
    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

# ------------------------------
# LSTM Model for Multi-Output Prediction
# ------------------------------
# Purpose: Define an LSTM model to predict both invocation counts and BlobBytes.
class LSTMModel(nn.Module):
    def __init__(self, input_size=7, hidden_size=128, num_layers=3, output_size=2):
        super(LSTMModel, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True, dropout=0.2)
        self.fc = nn.Linear(hidden_size, output_size)  # Output: count, BlobBytes
    
    def forward(self, x):
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)
        out, _ = self.lstm(x, (h0, c0))
        out = self.fc(out[:, -1, :])
        return out

# ------------------------------
# Training Function with Early Stopping
# ------------------------------
# Purpose: Train the LSTM with anti-overfitting measures (dropout, L2, early stopping).
def train_lstm(model, train_loader, test_loader, criterion, optimizer, num_epochs=15, patience=3, device='cuda' if torch.cuda.is_available() else 'cpu'):
    model = model.to(device)
    train_losses, val_losses = [], []
    best_val_loss = float('inf')
    epochs_no_improve = 0
    
    for epoch in range(num_epochs):
        # Train
        model.train()
        train_loss = 0
        for X, y in train_loader:
            X, y = X.to(device), y.to(device)
            optimizer.zero_grad()
            outputs = model(X)
            loss = criterion(outputs, y)  # Multi-output loss
            loss.backward()
            optimizer.step()
            train_loss += loss.item()
        train_losses.append(train_loss / len(train_loader))
        
        # Validate
        model.eval()
        val_loss = 0
        with torch.no_grad():
            for X, y in test_loader:
                X, y = X.to(device), y.to(device)
                outputs = model(X)
                loss = criterion(outputs, y)
                val_loss += loss.item()
        val_losses.append(val_loss / len(test_loader))
        
        print(f'Epoch [{epoch+1}/{num_epochs}], Train Loss: {train_losses[-1]:.4f}, Val Loss: {val_losses[-1]:.4f}')
        
        # Early stopping
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            epochs_no_improve = 0
            torch.save(model.state_dict(), '/kaggle/working/best_lstm_model.pth')
        else:
            epochs_no_improve += 1
            if epochs_no_improve >= patience:
                print(f"Early stopping at epoch {epoch+1}")
                break
    
    return train_losses, val_losses

# ------------------------------
# Binning Function for Classification Metrics
# ------------------------------
# Purpose: Bin invocation counts and BlobBytes for classification metrics.
def bin_values(values, thresholds=[10, 50]):
    """
    Bin values into low (0), medium (1), high (2) classes.
    thresholds: [low-to-medium, medium-to-high]
    """
    labels = np.zeros(len(values), dtype=int)
    labels[(values >= thresholds[0]) & (values < thresholds[1])] = 1
    labels[values >= thresholds[1]] = 2
    return labels

# ------------------------------
# Compute Metrics on Test Set
# ------------------------------
# Purpose: Calculate regression and classification metrics for count and BlobBytes on the test set.
def compute_metrics(y_test, y_pred, y_test_original, y_pred_original):
    metrics = {}
    for i, target in enumerate(['count', 'BlobBytes']):
        # Regression metrics
        r2 = r2_score(y_test_original[:, i], y_pred_original[:, i])
        mae = mean_absolute_error(y_test_original[:, i], y_pred_original[:, i])
        mse = mean_squared_error(y_test_original[:, i], y_pred_original[:, i])
        print(f"\nTest Set Regression Metrics ({target}):")
        print(f"R Score: {r2:.4f}")
        print(f"Mean Absolute Error (MAE): {mae:.4f}")
        print(f"Mean Squared Error (MSE): {mse:.4f}")
        metrics[f'{target}_r2'] = r2
        metrics[f'{target}_mae'] = mae
        metrics[f'{target}_mse'] = mse
        
        # Classification metrics
        thresholds = [10, 50] if target == 'count' else [100000, 1000000]  # Adjust for BlobBytes
        y_test_classes = bin_values(y_test_original[:, i], thresholds)
        y_pred_classes = bin_values(y_pred_original[:, i], thresholds)
        
        precision = precision_score(y_test_classes, y_pred_classes, average='weighted', zero_division=0)
        recall = recall_score(y_test_classes, y_pred_classes, average='weighted', zero_division=0)
        f1 = f1_score(y_test_classes, y_pred_classes, average='weighted', zero_division=0)
        
        print(f"\nTest Set Classification Metrics ({target}):")
        print(f"Precision (weighted): {precision:.4f}")
        print(f"Recall (weighted): {recall:.4f}")
        print(f"F1 Score (weighted): {f1:.4f}")
        metrics[f'{target}_precision'] = precision
        metrics[f'{target}_recall'] = recall
        metrics[f'{target}_f1'] = f1
        
        # Confusion matrix
        cm = confusion_matrix(y_test_classes, y_pred_classes)
        plt.figure(figsize=(6, 4))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Low', 'Medium', 'High'], yticklabels=['Low', 'Medium', 'High'])
        plt.title(f'Confusion Matrix for {target} (Test Set)')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.show()
    
    return metrics

# ------------------------------
# Main Execution
# ------------------------------
# Purpose: Train and evaluate LSTM models for multiple AnonAppNames, incorporating behavioral modeling.
if __name__ == "__main__":
    # Select top functions for training
    top_apps = df['AnonAppName'].value_counts().head(3).index  # Limit to 3 for speed
    print(f"Training models for functions: {top_apps.tolist()}")
    
    for function_id in top_apps:
        print(f"\nTraining LSTM for function: {function_id}")
        
        # Preprocess data
        X_train, y_train, X_test, y_test, scaler = preprocess_data(df, function_id=function_id, sequence_length=60)
        print(f"Train shape: {X_train.shape}, {y_train.shape}")
        print(f"Test shape: {X_test.shape}, {y_test.shape}")
        
        # Create datasets and dataloaders
        train_dataset = WorkloadDataset(X_train, y_train)
        test_dataset = WorkloadDataset(X_test, y_test)
        train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=32, shuffle=True)
        test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=32, shuffle=False)
        
        # Initialize model, criterion, optimizer
        model = LSTMModel(input_size=7, hidden_size=256, num_layers=4, output_size=2)
        criterion = nn.MSELoss()
        optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
        
        # Train model
        train_losses, val_losses = train_lstm(model, train_loader, test_loader, criterion, optimizer, num_epochs=20, patience=3)
        
        # Save model and scaler
        torch.save(model.state_dict(), f'/kaggle/working/lstm_model_{function_id}.pth')
        joblib.dump(scaler, f'/kaggle/working/scaler_{function_id}.pkl')
        print(f"Model saved to /kaggle/working/lstm_model_{function_id}.pth")
        
        # Plot loss
        plt.figure(figsize=(8, 4))
        plt.plot(train_losses, label='Train Loss')
        plt.plot(val_losses, label='Validation Loss')
        plt.title(f'LSTM Training and Validation Loss for {function_id}')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()
        plt.show()
        
        # Test predictions
        model.eval()
        device = 'cuda' if torch.cuda.is_available() else 'cpu'
        X_test_tensor = torch.tensor(X_test, dtype=torch.float32).to(device)
        with torch.no_grad():
            y_pred = model(X_test_tensor).cpu().numpy()
        y_pred_original = scaler.inverse_transform(np.hstack([y_pred, np.zeros((y_pred.shape[0], 5))]))[:, :2]
        y_test_original = scaler.inverse_transform(np.hstack([y_test, np.zeros((y_test.shape[0], 5))]))[:, :2]
        
        # Compute metrics
        metrics = compute_metrics(y_test, y_pred, y_test_original, y_pred_original)
        
        # Plot actual vs predicted for both targets
        for i, target in enumerate(['count', 'BlobBytes']):
            plt.figure(figsize=(12, 4))
            plt.plot(y_test_original[:200, i], label=f'Actual {target} (Test Set)')
            plt.plot(y_pred_original[:200, i], label=f'Predicted {target} (Test Set)')
            plt.title(f'Actual vs Predicted {target} for {function_id} (Test Set)')
            plt.xlabel('Time Step')
            plt.ylabel(target)
            plt.legend()
            plt.show()
        
        print(f"Sample prediction for {function_id} (count, BlobBytes): ({y_pred_original[-1, 0]:.2f}, {y_pred_original[-1, 1]:.2f})")
    
    # Behavioral Modeling by AnonUserId (example)
    print("\nBehavioral Modeling by AnonUserId:")
    user_groups = df.groupby('AnonUserId').agg({
        'AnonFunctionInvocationId': 'count',
        'BlobBytes': 'sum',
        'Read': 'sum',
        'Write': 'sum'
    }).rename(columns={'AnonFunctionInvocationId': 'count'})
    print(user_groups.describe())
````
